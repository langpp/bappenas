```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
from matplotlib import pyplot as plt
from matplotlib import style
import os
from fpdf import FPDF
import io
import base64

def makeChart(host, username, password, db, port, table, judul, columns, filePath, name, subjudul):
    try:
        connection = psycopg2.connect(user=username,password=password,host=host,port=port,database=db)
        cursor = connection.cursor()
        postgreSQL_select_Query = "SELECT * FROM "+table+" ORDER BY tanggal ASC"
    
        cursor.execute(postgreSQL_select_Query)
        mobile_records = cursor.fetchall() 
        uid = []
        lengthx = []
        lengthy = []
        for row in mobile_records:
            uid.append(row[0])
            lengthx.append(row[1])
            lengthy.append(row[2])
        #bar
        style.use('ggplot')
   
        fig, ax = plt.subplots()
    
        ax.bar(uid, lengthy, align='center')
        ax.set_title(judul)
        ax.set_ylabel('Total')
        ax.set_xlabel('Tanggal')
    
        ax.set_xticks(uid)
        ax.set_xticklabels((lengthx))
        b = io.BytesIO()
        plt.savefig(b, format='png', bbox_inches="tight")
        
        barChart = base64.b64encode(b.getvalue()).decode("utf-8").replace("\n", "")
        plt.show()
        
        #line
        plt.plot(lengthx, lengthy)
        plt.xlabel('Tanggal')
        plt.ylabel('Total')
    
        plt.title(judul)
        plt.grid(True)
        l = io.BytesIO()
        plt.savefig(l, format='png', bbox_inches="tight")
        
        lineChart = base64.b64encode(l.getvalue()).decode("utf-8").replace("\n", "")
        plt.show()
        
        #pie
        plt.title(judul)
        plt.pie(lengthy, labels=lengthx, autopct='%1.1f%%', 
        shadow=True, startangle=90)
    
        plt.axis('equal')
        p = io.BytesIO()
        plt.savefig(p, format='png', bbox_inches="tight")
        
        pieChart = base64.b64encode(p.getvalue()).decode("utf-8").replace("\n", "")
        plt.show()
        
        row = pd.read_csv(
            os.path.abspath(filePath),
            names=columns
        )
        
        del row['nama']
        del row['kategori']
        del row['jenis']
        del row['satuan']
        
        header = pd.read_csv(
            os.path.abspath(filePath),
            names=columns
        )
        
        del header['tanggal']
        del header['total']
        makePDF(row, header, judul, barChart, lineChart, pieChart, name, subjudul)        
        
    except (Exception, psycopg2.Error) as error :
        print (error)

    finally:
        if(connection):
            cursor.close()
            connection.close()
        
def uploadToPSQL(columns, table, filePath, engine):
    df = pd.read_csv(
        os.path.abspath(filePath),
        names=columns
    )
    
    del df['nama']
    del df['kategori']
    del df['jenis']
    del df['satuan']
    
    df.to_sql(
        table, 
        engine,
        if_exists='replace'
    )

    if len(df) == 0:
        return False
    else:
        return True
    
def makePDF(datarow, dataheader, judul, bar, line, pie, name, subjudul):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font('helvetica', 'B', 20.0)
    pdf.set_xy(100.0, 18.0)
    pdf.cell(ln=0, h=2.0, align='C', w=10.0, txt=judul, border=0)
    
    pdf.set_font('arial', '', 14.0)
    pdf.set_xy(100.0, 25.0)
    pdf.cell(ln=0, h=2.0, align='C', w=10.0, txt=subjudul, border=0)
    pdf.line(10.0, 30.0, 200.0, 30.0)
    pdf.set_font('times', '', 10.0)
    pdf.set_xy(17.0, 37.0)
    
    pdf.set_font('Times','',10.0) 
    datahead=list(dataheader.values)
    epw1 = pdf.w - 2*pdf.l_margin
    col_width1 = epw1/len(datahead[0])
    pdf.set_font('Times','B',10.0) 
    pdf.ln(0.5)
    
    th1 = pdf.font_size
    pdf.cell(col_width1, 2*th1, datahead[0][0], border=1, align='C')
    pdf.cell(col_width1, 2*th1, datahead[0][1], border=1, align='C')
    pdf.cell(col_width1, 2*th1, datahead[0][2], border=1, align='C')
    pdf.cell(col_width1, 2*th1, datahead[0][3], border=1, align='C')
    pdf.ln(2*th1)
    
    pdf.set_xy(17.0, 50.0)
    pdf.set_font('Times','',10.0) 
    data=list(datarow.values)
    epw = pdf.w - 2*pdf.l_margin
    col_width = epw/len(data[0])
    
    pdf.set_font('Times','B',10.0) 
    pdf.ln(0.5)
    
    th = pdf.font_size
    pdf.cell(col_width, 2*th, "Tanggal", border=1, align='C')
    pdf.cell(col_width, 2*th, "Total", border=1, align='C')
    pdf.set_font('Arial','',9)
    pdf.cell(epw, 2*th)
    for row in data:
        for datum in row:
            pdf.cell(col_width, 2*th, str(datum), border=1, align='C')
        pdf.ln(2*th)
    
    bardata = base64.b64decode(bar)
    barname = 'C:/Users/ASUS/Documents/bappenas/bloomberg/SektorEksternal/img/'+name+'-bar.png'
    with open(barname, 'wb') as f:
        f.write(bardata)
    
    linedata = base64.b64decode(line)
    linename = 'C:/Users/ASUS/Documents/bappenas/bloomberg/SektorEksternal/img/'+name+'-line.png'
    with open(linename, 'wb') as f:
        f.write(linedata)
        
    piedata = base64.b64decode(pie)
    piename = 'C:/Users/ASUS/Documents/bappenas/bloomberg/SektorEksternal/img/'+name+'-pie.png'
    with open(piename, 'wb') as f:
        f.write(piedata)
    
    pdf.add_page()
    col = pdf.w - 2*pdf.l_margin
    pdf.image(barname, link='', type='',x=10, y=15, w=col)
    pdf.add_page()
    col = pdf.w - 2*pdf.l_margin
    pdf.image(linename, link='', type='',x=10, y=15, w=col)
    pdf.add_page()
    col = pdf.w - 2*pdf.l_margin
    pdf.image(piename, link='', type='',x=10, y=15, w=col)
    pdf.ln(2*th)
    
    pdf.output('C:/Users/ASUS/Documents/bappenas/bloomberg/SektorEksternal/pdf/'+name+'.pdf', 'F')
    
columns = [
    "nama",
    "kategori",
    "tanggal",
    "total",
    "jenis",
    "satuan",
]

name = "SektorEksternal1_1"
host = "localhost"
username = "postgres"
password = "1234567890"
port = "5432"
database = "bloomberg_sektoreksternal"
table = name.lower()
judul = "Data Sektor Eksternal"
subjudul = "Badan Perencanaan Pembangunan Nasional"
filePath = 'C:/Users/ASUS/Documents/bappenas/bloomberg/SektorEksternal/'+name+'.csv';
engine = create_engine('postgresql://'+username+':'+password+'@'+host+':'+port+'/'+database)

checkUpload = uploadToPSQL(columns, table, filePath, engine)
if checkUpload == True:
    makeChart(host, username, password, database, port, table, judul, columns, filePath, name, subjudul)
else:
    print("Error When Upload CSV")
```
